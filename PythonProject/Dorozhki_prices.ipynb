```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("norm_dorozhki.xlsx")
df['Цена'] = df['Цена'].str.replace('₽', '').str.replace(' ', '').astype(int)
plt.figure(figsize=(10, 6), facecolor='#FFF0F5')
sns.histplot(df['Цена'], bins=10, color='hotpink', kde=True)

plt.title('Распределение цен на беговые дорожки', fontsize=16, color='deeppink')
plt.xlabel('Цена (₽)', fontsize=14, color='deeppink')
plt.ylabel('Количество', fontsize=14, color='deeppink')
plt.xticks(color='deeppink')
plt.yticks(color='deeppink')
ax = plt.gca()
ax.set_facecolor('#FFF0F5')

plt.show()
```
