```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("norm_dorozhki.xlsx")
df['Цена'] = df['Цена'].str.replace('₽', '').str.replace(' ', '').astype(int)
df['Макс вес'] = df['Макс вес'].str.replace(' кг', '').astype(int)
df['Мощность двигателя'] = df['Двигатель'].str.extract(r'(\d+\.?\d*)').astype(float)


plt.figure(figsize=(12, 8), facecolor='#FFF0F5')

scatter = sns.scatterplot(
    x=df['Мощность двигателя'],
    y=df['Цена'],
    hue=df['Макс вес'],
    size=df['Макс вес'],
    sizes=(50, 300),
    palette='magma',
)


plt.title('Зависимость цены от мощности двигателя и максимального веса', fontsize=16, color='deeppink')
plt.xlabel('Мощность двигателя (л.с.)', fontsize=14, color='deeppink')
plt.ylabel('Цена (₽)', fontsize=14, color='deeppink')

for i, row in df.iterrows():
    plt.text(
        row['Мощность двигателя'],
        row['Цена'],
        row['Название'],
        fontsize=8,
        color='deeppink',
        alpha=0.7,
    )

plt.xticks(color='deeppink')
plt.yticks(color='deeppink')

legend = plt.legend(title='Макс вес (кг)', title_fontsize=12, fontsize=10, facecolor='#FFF0F5', edgecolor='deeppink')
for text in legend.get_texts():
    text.set_color('deeppink')
ax = plt.gca()
ax.set_facecolor('#FFF0F5')
plt.show()
```
